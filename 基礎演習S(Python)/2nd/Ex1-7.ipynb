```python
# [1] urlから講義内容を
library(magrittr)
library(rvest)
getstxt <- function(url){
  dat <- read_html(url)			# 読み込み
  XPath <- "//tr[1]/td[3]/table/tr/td"				# 「講義内容」部分のXPath
  lec <- html_nodes(dat, xpath=XPath) %>% html_text(lec) %>% gsub("\\r|\\n| |\\t","",lec)				# 無駄な文字列（空白文字）の削除　※英文等では問題
  return(lec)
}
```

```python
# [2]
savestxt <- function(ids,urls,sdir){
  for(u in length(ids)){
    writeLines(getstxt(urls[u]), sdir)
  }
}
```
