```python
# 関数定義
x=c(1, 2, 3, 4, 5)
svar<-function(x){
  m<-mean(x)
  d<-x-m
  d2<-d^2
  ss<-sum(d2)
  n<-length(x)
  res<-ss/n
  return(res)
}
#-----------------------------------
# 使用例
svar(x)
#-----------------------------------
# 参考
sum((x-mean(x))^2)/length(x)    # 簡約版(1)
var(x)*(length(x)-1)/length(x)  # 簡約版(2)
```
