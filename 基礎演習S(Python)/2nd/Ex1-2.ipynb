### Ex.1-2　「食べログ」からのデータ取得

#####[1]

```python
library(rvest)
html <- read_html("https://tabelog.com/tokyo/A1310/A131002/13167573/") # HTML読み込み
html	# 確認　※DOM（Document Object Model）形式で表示
tmp <- html_nodes(html, xpath="//a[@class='rdheader-budget__price-target']")  # HTMLの要素抽出
tmp					# 確認（2要素）
price <- html_text(tmp[1])			# htmlから属性，テキスト，タグ名を抽出。[1]が夕食に該当するので以下では、tmp[1]を使用。
price					# 確認(夕食価格である)
price1 <- price %>% gsub(pattern="^.{8}",replacement="") %>%
  gsub(pattern = ",",replacement = "")  #文頭から八文字削除し、その後コンマを除けば、求めたい値となる。
price1					# 確認(求めたい値が出力された)
```
