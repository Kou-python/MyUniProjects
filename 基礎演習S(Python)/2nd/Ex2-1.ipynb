```python
# [1] fn：文章ファイル → 形態素
s <- "すもももももももものうち"    # 文字列を変数ｓに代入
fn <- "tmp.txt"    # 文字列代入
write(s, fn)    # tmp.txtという名前のファイルを生成し、そこへ変数ｓを出力
dir()    # 現在ディレクトリのファイル一覧を出力
readLines(fn)    # 変数fnを読み取った文字列を出力する
# G.Colab.の左の「ファイル」メニューを開き「更新」アイコンを押して作ったファイルを確認せよ

res1 <- docDF(fn, type=1);    #変数fn(txtファイル) を単語ごとに分割し、その出現頻度をdataframeで返す
print(res1)    # 文字列で変数res1を出力
res0 <- docDF(fn, type=0); res0    # 参考
```

```python
# [2] d：データフレーム（の列要素）→ 形態素
ID <- 1:3    # 1から3までの数値を変数IDへ代入する
Sex <- c("F", "M", "F")    # 3つの文字列をSexへ代入する
Rem <- c("今朝は早く起きた。", "昨夜は遅く寝た。", "今朝の食事は遅かった．")    # 3つの文字列をRemへ代入する
d <- data.frame(ID, Sex, Rem)    # 上の3つの変数をdataframe型に落とし込む
d    #  dを出力

res <- docDF(d, 3, type=1); res    # 変数dの3列目の文字列について
res <- docDF(d, 3, type=1, pos=c("名詞","形容詞")); res    #posで名詞と形容詞だけに出力が限定される
res <- docDF(d, 3, type=1, pos=c("名詞","形容詞"), minFreq=2); res    #2行目で出力された表のrow1~row3で最低2回以上出現する行のみを出力する
```

```python
# [3]　dn：（文章ファイルの）ディレクトリ → 形態素（文書ターム行列）
dn <- "tmpd"    # 文字列を代入
dir.create(dn)    # 変数dnのディレクトリを作成
fn <- c("f1.txt", "f2.txt", "f3.txt")    # 3つの文字列を変数fnへ代入する
fn <- paste0(dn, "/", fn)    # fnをtmpd/f_.txt(_には1から順に入る)に文字列を更新する
for (i in 1:3){    # 3回ループ文
  write(Rem[i], fn[i])    # Rem[i]をディレクトリfn[i]へ書き込む
}    # 上へ戻る
dir(dn)    # dnのディレクトリ内ファイルを確認する
# G.Colab.の左の「ファイル」メニューを開き「更新」アイコンを押して作ったディレクトリを確認せよ

res <- docDF(dn, type=1); res    #変数dnのディレクトリにあるtxtファイル3つに関して形態素解析を行う
res <- docDF(dn, type=1, minFreq = 2); res    #上の行の条件に2つ以上の文でマッチしたものを出力対象とする
res <- docDF(dn, type=1, pos=c("名詞", "動詞")); res    #1行目の条件にposが名詞と動詞のみに絞られたものを出力対象とする
```

```python
# [4]　dn：ディレクトリ → 形態素（文書ターム行列）※重み付き　
# 参照：http://www.ic.daito.ac.jp/~mizutani/mining/term_document_matrix.html
# 「検索語・文書行列の重み付け 〜 TF・IDF法」と
# 「TF・IDF法による検索語・文書行列の規格化」の部分を読むこと
res <- docDF(dn, type=1, pos=c("名詞", "形容詞", "動詞"), weight="tf*idf"); res
    # 変数dnのディレクトリにあるtxtファイル3つに関して形態素解析を行う。内容はposに名詞・形容詞・動詞を含むもの。<dbl>にはweightの値が出力される。
res <- docDF(dn, type=1, pos=c("名詞", "形容詞", "動詞"), weight="tf*idf*norm"); res
    # 変数dn
res[,4:6]    # resの4:6列目を出力
colSums(res[,4:6]^2)  # 列の合計の2乗を出力する。
```

```python
# [5] 　dn：ディレクトリ（or fn：ファイル） → N-gram
# 参考：「Nグラム統計」http://www.ic.daito.ac.jp/~mizutani/mining/n_gram.html
# に目を通すこと
dn <- "tmpd"    # dnに文字列tmpdを代入する
res <- NgramDF2(dn, type=1, N=2); res    # resにNgram関数（dnの2グラム、言語単位=形態素を代入
res <- NgramDF2(dn, type=1, N=2, pos=c("名詞", "動詞")); res    # resにNgram関数（dnの名詞・動詞からなる2グラム、言語単位=形態素を代入
res <- NgramDF2(dn, type=1, N=3, pos=c("名詞", "形容詞", "動詞")); res    # resにNgram関数（dnの名詞・動詞・形容詞からなる3グラム、言語単位=形態素を代入
res <- NgramDF2(dn, type=2, N=2); res    # resにNgram関数（dnの2グラム、言語単位=品詞を代入
fn <- "tmpd/f2.txt"    # fnに文字列（ファイルパス）を代入
res <- NgramDF2(fn, type=0, N=2); res    # resにNgram関数（fnの2グラム、言語単位=文字を代入
```

```python
# [6] コロケーションとTスコア，MIスコア
# 参考：「コロケーションと語彙分析」
# http://www.ic.daito.ac.jp/~mizutani/mining/collocation.html
# に目を通すこと 
fn <- "tmpd/f3.txt"    # fnにファイルパスを代入
readLines(fn)    # fnの文字列を出力
collocate(fn, node="は", span=1)    # 変数fnのファイルの中心語「は」、span値1として関数collocateを実行する
collocate(fn, node="は", span=2)    # 変数fnのファイルの中心語「は」、span値2として関数collocateを実行する
collocate(fn, node="は", span=3)    # 変数fnのファイルの中心語「は」、span値3として関数collocateを実行する

res <- collocate(fn, node="は", span=2)    # resに変数fnのファイルの中心語「は」、span値2として関数collocateの実行を代入する
collScores(res, node="は", span=2)    # resのTscore及びMIscoreを出力した
```
