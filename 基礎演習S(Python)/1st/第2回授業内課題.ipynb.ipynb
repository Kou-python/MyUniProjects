#問1

```python
a <- dat $ 科目A
b <- rep("", length(a))		# 空のベクトルを用意
for(i in 1:length(a)){
  if(a[i]<=10)b[i]<-"0-10" else
  if(a[i]<=20)b[i]<-"10-20" else
  if(a[i]<=30)b[i]<-"20-30" else
  if(a[i]<=40)b[i]<-"30-40" else
  if(a[i]<=50)b[i]<-"40-50" else
  if(a[i]<=60)b[i]<-"50-60" else
  if(a[i]<=70)b[i]<-"60-70" else
  if(a[i]<=80)b[i]<-"70-80" else
  if(a[i]<=90)b[i]<-"80-90" else
  b[i]<-"90-100" 
}
table(b)
```
