```python
標本分散<-numeric(10000)
for(i in 1:10000){
  r<-rnorm(n=10, mean=50, sd=10)
  標本分散[i]<-var(r)*(length(r)-1)/length(r)
}
hist(標本分散)
```

#2

```python
不偏分散<-numeric(10000)
for(i in 1:10000){
  r<-rnorm(n=10, mean=50, sd=10)
  不偏分散[i]<-var(r)
}
hist(不偏分散)
```
