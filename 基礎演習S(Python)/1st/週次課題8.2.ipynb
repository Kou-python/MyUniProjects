• respon.csvはみなさんの授業内課題5.3の解答をresponで集めたものです。
> respon <- read.csv(“respon.csv”, header=T)
• このデータを用いて全員の結果からヒストグラムを描きたいと思います。
• ところが、解答の中にはデータだけでなく、それを求めるための式やRの出力をそのままコピーした
もの等が含まれ、このままではヒストグラムは描けません。
• grep、sub、gsub等を用いて、必要な数値のみを抜き出して下さい。(1回で上手く行かなくても、
何回かこれらの関数を用いることで、ヒストグラムを作成するのに必要なデータが求まるでしょ
う。)
• 最後に作成したデータを用いてヒストグラムを描いて下さい。(40以上60以下に収まらない値は除
いてヒストグラムを描くこと。breaksも程よく設定して下さい。)

```python
respon <- read.csv("/respon.csv", header=T)[,1]
data<- as.numeric(sub("^(?:.|\\n)*?(\\d{2}\\.\\d+)(?:.|\\n)*?$", "\\1", respon, perl = T))
data
hist(data[data>40&data<60],breaks = seq(40,60,0.5))
```
