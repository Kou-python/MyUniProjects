```python
dat<-read.csv("/web2")[,1]
```

#1 a i u e o をこの順に含む単語は?

```python
grep("a.*i.*u.*e.*o", dat, perl = T, value = T)
```

#2 a を5つ以上含む単語は?(量化子を使用すること)

```python
grep("(?:a.*){5,}", dat, perl = T, value = T)
```

#3 a または A で始まり z で終わる単語は?

```python
grep("^[aA].*z$", dat, perl=T, value = T)

#(?:a|A)=[aA]
```

#4 同じ文字が3文字以上連続して現れる単語は?

```python
grep("(.+)\\1{2,}",dat, perl = T, value = T)
```
