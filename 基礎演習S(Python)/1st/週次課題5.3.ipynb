##2
サイコロの出る目の確率をそれぞれ1/21,2/21,3/21,4/21,5/21,6/21とした。

```python
y=numeric(10000)
for(i in 1:10000){
  dice2<- sample(1:6, size = 10, replace = TRUE, 
  prob = c(1/21,2/21,3/21,4/21,5/21,6/21))
  y[i]=mean(dice2)
}
hist(y)
```
