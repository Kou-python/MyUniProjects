### 1. 散布図

```python
dat<-read.csv("/score2022S.csv")
A<-dat$`科目A`
B<-dat$`科目B`

plot(A,B, col="#ff8c00", main="A,B sanpuzu")
```

### 2. 共分散、相関係数

```python
print("共分散")
cov(A,B)
print("相関係数")
cor(A,B)
```

### 3.クロス表

```python
科目a<-cut(A,c(-1,59,100))
levels(科目a)<-c("不合格","合格")
科目b<-cut(B,c(-1,59,100))
levels(科目b)<-c("不合格","合格")

table(科目a,科目b)
```
