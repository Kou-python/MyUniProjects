#1.実際にサイコロを2回だけ振って、その目の平均が3.5になる確率はどのくらいでしょうか。

```python
cnt=0
for(i in 1:10000){
  cal<-ceiling(runif(2,0,6))
  if(mean(cal)==3.5)cnt=cnt+1
}
cnt/10000
```

1万回中何回平均値が3.5になったのかをifで範囲指定して計算した。

#2.サイコロを振る回数を4回、10回、100回、10000回と増やすと、その目の平均が3.5±0.01の範囲に収まる確率はどのくらいになるでしょうか。

それぞれ以下の回数サイコロを振り、その試行を1万回実行し、その目の平均が3.5±0.01の範囲に収まる時にifで範囲指定し、そこでcnt変数を加算していき、最後にその値を1万で割ることで、その目の平均が3.5±0.01の範囲に収まる確率を求めた。

サイコロを4回振った時

```python
#4回フル
cnt=0
num=4
for(i in 1:10000){
  cal<-ceiling(runif(num,0,6))
  if(abs(mean(cal)-3.5)<00.1) cnt=cnt+1
}
cnt/10000
```

サイコロを10回振った時

```python
#10回フル
cnt=0
num=10
for(i in 1:10000){
  cal<-ceiling(runif(num,0,6))
  if(3.49<=mean(cal)&mean(cal)<=3.51)cnt=cnt+1
}
cnt/10000
```
