```python
score <- read.csv("/score2022S.csv", header=T)
変数IDs<-score[,1]
```

```python
grep("3",変数IDs,perl = T,value=T)#間違えて "3+"にしてた
```

```python
grep("E$",変数IDs,perl=T,value = T)
```

```python
grep("-.1...",変数IDs,perl = T,value = T)
```

```python
#grep("1[3,]",変数IDs,perl = T,value = T)　間違えた
grep("1.*1.*1.*", 変数IDs, perl = T, value = T)
```
